```python
print("Jumlah transaksi oleh seluruh penumpang pensiun:",jmlpensiun)
print("Jumlah transaksi oleh penumpang perempuan pensiun:", female)
print("Jumlah transaksi oleh penumpang laki-laki pensiun:", male)
```

# Soal No 4

# Soal No 5

Berapakah biaya yang dikeluarkan dan rute perjalanan untuk transaksi dengan jarak tempuh terlama dan jumlah pemberhentian terbanyak?


# Convert Time Column

```python
df_time = df.copy()

df_time['tapInTime'] = pd.to_datetime(df_time['tapInTime'])
df_time['tapOutTime'] = pd.to_datetime(df_time['tapOutTime'])
```

# Extraction

```python
df_time['time_only_In'] = df_time['tapInTime'].dt.time
df_time['time_only_Out'] = df_time['tapOutTime'].dt.time
```

# Calculation

```python
df_time['paling_lama'] = df_time['tapOutTime'] - df_time['tapInTime']
df_time['total_stop'] = df_time['stopEndSeq'] - df_time['stopStartSeq']
```

# Sequencing

```python
df_terlama = df_time.sort_values(by=['paling_lama','total_stop'],ascending=[False,False])
```
